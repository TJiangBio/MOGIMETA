```python
import numpy as np
import os
import pandas as pd
#这个代码是可解释性相关的
#我们使用了三个特征去训练模型，这里要看哪个特征更起作用
```

```python
path = "E://2022mofometa//test_result//node_weight"
file_list = os.listdir(path)
P_file_list = file_list[0:377]
N_file_list = file_list[377:]
```

```python
max_sub_graph = pd.read_csv('utils_files//max_sub_graph.csv')
```

```python
#Positive patients
P_node_weight = []
for i in P_file_list:
    #node_w = np.loadtxt(path+"//"+i)
    data = np.array(pd.read_table(path+"//"+i,sep=' ',header=None))
    P_node_weight.append(data)
    
#Positive patients
P_N_W = np.concatenate(P_node_weight, axis=1)
pnw = P_N_W.mean(axis=1)
P_sort_node_weight = np.argsort(pnw)

P_sort_node_weight_list = list(P_sort_node_weight)
P_sort_nodes = max_sub_graph.iloc[P_sort_node_weight_list]


```

```python
P_sorted_weights = np.sort(pnw)
```

```python
P_sorted_weights = pd.DataFrame(P_sorted_weights)
```

```python
path_P_nodes_weights = "E://2022mofometa//test_result//P_nodes_weights"
P_sorted_weights.to_csv(path_P_nodes_weights+".csv",index=False)
```

```python
P_sorted_nodes = pd.concat([P_sort_nodes['0'], P_sorted_weights], axis=1)
```

```python
#Positive patients
path_P_nodes = "E://2022mofometa//test_result//P_nodes"
P_sorted_nodes.to_csv(path_P_nodes+".csv",index=False)
```

```python
#Negative patients
N_node_weight = []
for i in N_file_list:
    #node_w = np.loadtxt(path+"//"+i)
    data = np.array(pd.read_table(path+"//"+i,sep=' ',header=None))
    N_node_weight.append(data)
    
#Negative patients
N_N_W = np.concatenate(N_node_weight, axis=1)
nnw = N_N_W.mean(axis=1)
N_sort_node_weight = np.argsort(nnw)
N_sort_node_weight_list = list(N_sort_node_weight)
N_sort_nodes = max_sub_graph.iloc[N_sort_node_weight_list]
```

```python
N_sorted_weights = np.sort(nnw)
```

```python
N_sorted_weights = pd.DataFrame(N_sorted_weights)
```

```python
path_N_nodes_weights = "E://2022mofometa//test_result//N_nodes_weights"
N_sorted_weights.to_csv(path_N_nodes_weights+".csv",index=False)
```

```python
N_sorted_nodes = pd.concat([N_sort_nodes['0'], N_sorted_weights], axis=1)
#Negative patients
path_N_nodes = "E://2022mofometa//test_result//N_nodes"
N_sorted_nodes.to_csv(path_N_nodes+".csv",index=False)
```
